```python
n = 1000
rollCount = 0
historyD2 = []
historyD3 = []
# for uitkomst in throws:
#     historyD2.append(uitkomst)


while rollCount < n:
    d2.roll()
    rollD2 = d2.getNumber()
    historyD2.append(rollD2)
    np.append(histD2, rollD2)
    d3.roll()
    rollD3 = d3.getNumber()
    historyD3.append(rollD3)
    np.append(histD3, rollD3)
    rollCount += 1

histD2 = d2.getRollHistory(historyD2)
histD3 = d3.getRollHistory(historyD3)
print(len(histD2))
print(len(histD3))
print(histD2)
print(histD3)
# d2.roll()
# taco = d2.getNumber()
# print(taco)
```
